```python
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)
print(message.content[0].text)


message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role":"user",
            "content":"Write another sentence"
        }
    ]

)
print(message.content[0].text)
```

```python
def append_user_message(messages, text):
    msg = {"role":"user","content":text}
    messages.append(msg)

def append_assistant_message(messages, text):
    msg = {"role":"assistant","content":text}
    messages.append(msg)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return message.content[0].text
```

```python
# multi-turn conversations - each time a follow question is asked we need to send the whole conversation, 
# claude do not store the previous messages
conversation = []

append_user_message(conversation,"Define Quantum computing in one sentence")

response = chat(conversation)

print("answer: ", response)
# add claude response to the conversation history
append_assistant_message(conversation, response)

# add a follow up question to conversation
append_user_message(conversation, "Write another sentence")

final_answer = chat(conversation)

print("final answer: ",final_answer)
```

```python
def print_conversation(conversation:list):
    for message in conversation:
        if message["role"] == "user":
            print("> ", message["content"])
        else:
            print("------")
            print(message["content"])
            print("------")
    return
```

```python
# simple chat bot
conversation2 = []
while True:
    user_message = input("Enter your prompt: ")
    if user_message == "":
        break
    print("> ", user_message)
    append_user_message(conversation2, user_message)
    response = chat(conversation2)
    print("---------")
    print(response)
    print("---------\n")
    append_assistant_message(conversation2, response)
    print_conversation(conversation2)
```
